```python
import numpy as np
import random
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D,MaxPooling2D,Dense, Flatten
print("sucessfully imported")
```

```python
x_train=np.loadtxt('input.csv', delimiter=',')
y_train=np.loadtxt('labels.csv', delimiter=',')

x_test=np.loadtxt('input_test.csv', delimiter=',')
y_test=np.loadtxt('labels_test.csv', delimiter=',')

x_train=x_train.reshape(len(x_train),100,100,3)
y_train=y_train.reshape(len(y_train))

x_test=x_test.reshape(len(x_test),100,100,3)
y_test=y_test.reshape(len(y_test))

x_train=x_train/255.00
y_train=y_train/255.00
x_test=x_test/255.00
y_test=y_test/255.00
```

```python
print("Shape of x_train: ", y_train.shape)
```

```python
idx=random.randint(0,len(x_train))
plt.imshow(x_train[idx,:])
plt.show()
```

```python
#Model
```

```python
model=Sequential([
     Conv2D(32,(3,3),activation='relu', input_shape=(100,100,3)),
     MaxPooling2D((2,2)),

    Conv2D(32,(3,3),activation='relu'),
    MaxPooling2D((2,2)),

    Flatten(),
    Dense(64,activation='relu'),
    Dense(1,activation='sigmoid')

    
])
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(x_train,y_train, epochs=10,batch_size=64)
```

```python
model.evaluate(x_test,y_test)
```

#Making Prediction

```python
idx2=random.randint(0,len(y_test))
plt.imshow(x_test[idx2,:])
plt.show()

if (y_test[idx2]):
    print("The iamge is for Cat")

else:
    print("The image is for Dog")

y_pred=model.predict(x_test[idx2,:].reshape(1,100,100,3))
print(y_pred)

y_pred=y_pred>0.5
if y_pred:
    print("Model predict it as Cat")

else:
    print("Model predict it as Dog")
```
